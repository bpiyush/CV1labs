#### Varying $\theta$

```python
thetas = [0, np.pi/4, np.pi/2]
labels = ["$0$^{o}", "$45$^{o}", "$90$^{o}"]
gabors = []

for theta in thetas:
    gabor = createGabor(sigma=10, theta=theta, lamda=30, psi=0.0, gamma=0.25)
    gabors.append(gabor)
```

```python
fig, ax = plt.subplots(3, 2, figsize=(8, 12), constrained_layout=True)

for i in range(3):
    for j in range(2):
        _ax = ax[i][j]

        _ax.imshow(gabors[i][:, :, j])
        
        title = "Real part" if j == 0 else "Imaginary part"
        if i == 0:
            _ax.set_title(title, fontsize=22)
        
        if j == 0:
            _ax.set_ylabel("$ \\theta = {}$".format(labels[i]), fontsize=20)
        
        _ax.set_xticks([])
        _ax.set_yticks([])

plt.savefig("../results/gabor_varying_theta.png", bbox_inches="tight")
plt.show()
```
